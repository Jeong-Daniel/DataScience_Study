인터페이스는 상태를 저장하고 plt명령어가 적용되는 곳에 있는 현재 그림과 축즐 저장하고 plt.gcf를 사용해 이를 참조할 수 있다.

```python
plt.figure() # 플롯 그림을 완성
#두 개의 패널 중 첫 번째 패널을 생성하고 현재 축(aixs)을 설정
plt.subplot(2,1,1) #(row, columns, pane number)
plt.plot(x, np.sin(x))

#두 번째 패널을 생성하고 현재 축(aixs)을 설정
plt.subplot(2,1,2)
plt.plot(x, np.cos(x));
```

### 객체지향 인터페이스
객체지향 인터페이스에서 플로팅 함수는 '활성화된' 그림이나 축의 개념에 의존하지 않는 명시적인 Figure와 Axes 객체의 매서드다. 이 플로팅 스타일을 이용해 앞에서 본 플롯을 다시 만들려면 다음 코드를 실행하면 된다.

```python
# 먼저 플롯 그리드를 생성
# ax는 두 개의 축 객체의 배열이 됨
fig, ax = plt.subplots(2)

# 적절한 객체에서 plot() 매서드를 호출
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

### 간단한 라인 플롯

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

모든 Matplotlib 플롯은 그림figure과 축axes을 만드는 것으로 시작한다.

```python
fig = plt.figure()
ax = plt.axes()
```

matplotlib에서 그림은 축과 그래픽 텍스트 레이블을 표시하는 모든 객체를 포함하는 하나의 컨테이너로 생각하면 된다. 축은 위에서 본 것 처럼 눈금과 레이블이 있는 테두리 상조로 나중에 시각화를 형성하는 플릇 요소를 포함하게 된다. 이 책에서 변수 이름 fig는 그림 인스턴스, ax는 축 인스턴스를 말할 때 사용한다.

축을 만들고 나면 ax.plot함수를 사용해 일부 데이터를 플로팅할 수 있다. 간단한 사인 곡선으로 시작해본다.

```python
fig = plt.figure()
ax= plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x, np.sin(x))
#plt.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

### 플롯 수정하기: 선 색상과 스타일
plt.plot()함수는 추가 인수를 받는다.

```python
plt.plot(x, np.sin(x-0), color='blue') #색상을 이름으로 지정
plt.plot(x, np.sin(x-1), color='g') #짧은 색상 코드 rgbcmyk
plt.plot(x, np.sin(x-2), color='0.75') #0과 1사이로 회색조 지정
plt.plot(x, np.sin(x-3), color='#FFDD44') #16진수 코드(RRGGBB, 00~FF사이)
plt.plot(x, np.sin(x-4), color=(1.0,0.2,0.3)) #RGB튜플, 0과 1값
plt.plot(x, np.sin(x-5), color='chartreuse') #모든 HTML색상 이름을 지원
```

아무 색도 지정하지 않으면 Matplotlib이 여러 줄에 대해 자동으로 기본 색상 집합을 돌아가며 색상을 채택한다.  
이와 비슷하게 linestyle 키워드를 사용하면 선 스타일을 조정할 수 있다.

```python
plt.plot(x, x+0, linestyle='solid') # 축약형은 '-'
plt.plot(x, x+1, linestyle='dashed') # '--'
plt.plot(x, x+2, linestyle='dashdot') # '-.'
plt.plot(x, x+3, linestyle='dotted'); #':'
```

좀더 간결하게 지정하고 싶다면 linestyle과 color코드를 키워드가 없는 단일 인수로 결합해서 plt.plot()함수에 전달하면 된다.

```python
plt.plot(x,x+0, '-g') # solid green
plt.plot(x,x+1, '--c') # dashed cyan
plt.plot(x,x+2, '-.k') # dashodt black
plt.plot(x,x+3, ':r') # dotted red
```

단일 문자로 된 색상 코드는 디지털 색상 그래픽에 보편적으로 사용되는 RGB(Red, Green, Blue)와 CMYK(Cyan청록, Magenta 자홍, Yello노랑, Black 검정) 표준 약어를사용한다.

### 플롯 조정하기: 축 경계
축 경계를 조정하는 가장 기본적인 방식은 plt.xlim()과 plt.ylim()매서드를 사용하는 것이다.

```python
plt.plot(x, np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x, np.sin(x))
plt.xlim(10,0)
plt.ylim(1.2, -1.2)
```

이와 관련한 유용한 메서드는 plt.axis()다 (axes, axis를 혼동하지말자) plt.axis() 매서드는 [x축 최솟값, x축 최댓값, y축 최솟값, y축 최댓값]을 지정하는 리스트를 전달해 한 번의 호출로 x와 y축의 경계를 설정하게 해준다.

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5])
```

plt.axis() 매서드를 이용하면 이를 넘어서 현재 플롯 주변의 경계를 자동으로 더 밀착시키는 당의 작업도 할 수 있다.

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

이 메서드를 이용하면 좀 더 높은 단계의 지정도 가능해서 가로세로 비율을 균등하게 설정해 화면상에 x축 한 단위와 y축의 한 단위가 똑같에 설정하는 등의 작업을 할 수 있다.

```python
plt.plot(x,np.sin(x))
plt.axis('equal')
```

### 플롯에 레이블 붙이기
제목과 축은 가장 간단한 레이블이다.

```python
plt.plot(x, np.sin(x))
plt.title("A Sin Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

레이블의 위치와 크기, 스타일을 수정하려면 그 함수에 선택적인 인수를 사용하면 된다. 범례를 만들기 위해서는 plt.legend()를 사용한다.

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend()
```

 스타일과 객체지향 메서드 차이  
 - ``plt.xlabel()``  → ``ax.set_xlabel()``
- ``plt.ylabel()`` → ``ax.set_ylabel()``
- ``plt.xlim()``  → ``ax.set_xlim()``
- ``plt.ylim()`` → ``ax.set_ylim()``
- ``plt.title()`` → ``ax.set_title()``

플로팅을 위한 객체지향 인터페이스에서는 이 함수들을 개별적으로 호출하는 대신 ax.set()매서드를 사용해 한번에 이 모든 속성을 설정하는 것이 더 편리하다.

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
```
