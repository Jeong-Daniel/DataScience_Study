```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#회귀 알고리즘 구현
#y = beta0 + beta1 * x

def load_data():
    X = 2*np.random.rand(100,1)
    y = 6 + 4* X +np.random.rand(100,1)
    return X, y

def predcition(beta_0, beta_1, X):
    pred_y = beta_0 + beta_1 * X
    return pred_y
    
def update_beta(beta_0, beta_1, X, loss, lr):
    delta_0 = -(lr*(2/len(loss)) * (np.dot(X.T ,loss)))
    delta_1 = -(lr*(2/len(loss)) * np.sum(loss))
    return delta_0, delta_1

def plotting_graph(X,y,beta_0, beta_1):
    y_pred = beta_0 + beta_1 * X
    fig = plt.figure()
    plt.scatter(X,y)
    plt.plot(X,y_pred, c='r')
    plt.show()

def gradient_descent(X,y,iteration,lr):
    beta_0 = np.zeros((1,1))
    beta_1 = np.zeros((1,1))
    
    for i in range(iteration):
        loss = y - predcition(beta_0, beta_1, X)
        beta0_delta, beta1_delta = update_beta(beta_0, beta_1, X, loss, lr)
        beta_0 = beta0_delta
        beta_1 = beta1_delta
        if i % 1000 == 0:
            print("학습 횟수 : ", i)
            plotting_graph(X,y,beta_0, beta_1)
    return beta_0, beta_1


def main():
    lr = 1e-4
    iteration = 10000
    X, y = load_data()
    beta_0, beta_1 = gradient_descent(X,y,iteration,lr)
    print("{} 번의 학습 이후의 회귀 알고리즘 결과".format(iteration))
    print("beta_0", beta_0, "beta_1 ",beta_1)
    plotting_graph(X,y,beta_0,beta_1)
    
main()
```
